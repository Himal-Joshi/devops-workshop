```python
doc = """
DevOps and MLOps form the operational backbone of modern tech infrastructure, bridges built to connect software engineering and data science directly to automated production environments. While DevOps centers on the rapid, reliable deployment of software code through continuous integration and infrastructure-as-code tools like GitHub Actions, Docker, and Terraform, MLOps extends these principles to manage the highly volatile variables of data and machine learning models. Navigating their documentation means balancing foundational blueprints—such as roadmap.sh/devops for systemic infrastructure tracks and Google’s MLOps Architecture Guide for understanding pipeline automation maturity—with tool-specific ecosystems. For standard operations, mastering the official manuals for Kubernetes and Ansible ensures scalable application hosting and configuration, whereas the MLOps space shifts focus heavily toward lifecycle management platforms like MLflow and Weights & Biases for experiment tracking, alongside DVC (Data Version Control) to version massive training datasets in tandem with Git. Ultimately, reading through these documentation suites reveals that while DevOps aims to maintain high uptime and seamless code delivery, MLOps layers on the critical necessity of tracking data drift, validation schemas, and model decay over time.
"""
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

# Example 1: Set a smaller chunk_overlap
text_splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=200)
text_splitter.split_text(doc)

```

```python
# In your text splitting cell:
chunked_texts = text_splitter.split_text(doc)
len(chunked_texts)
```

```python
from langchain_ollama import OllamaEmbeddings

embedding_model = OllamaEmbeddings(model='nomic-embed-text:latest')
```

```python
from langchain_core.vectorstores import InMemoryVectorStore

vector_store = InMemoryVectorStore(embedding_model)

vector_store.add_texts(chunked_texts)
```

## Workflow 2

```python
prompt = "what is deployment"
```

```python
result=vector_store.similarity_search(query=prompt, k=1)
result
```

```python
final_prompt = """

{prompt}

{result[0].page_content}

Context : Summarize the response in few sentences in Nepali language.

"""

final_prompt
```

```python
from urllib3 import response
from langchain_ollama import ChatOllama

model = ChatOllama(model="qwen3.5:2b")
res = model.invoke(final_prompt)
print(response.content)
```
